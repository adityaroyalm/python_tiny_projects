```python
import pandas as pd
import seaborn.apionly as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
obj=pd.read_csv('C:/Users/aditya royal/Downloads/MPG.csv')
#plt.show()
mpg_year=obj.groupby(['model year']).mean()
plt.scatter(mpg_year.index,mpg_year['mpg'])
plt.plot(mpg_year.index,mpg_year['mpg'])
plt.xlabel('model_year')
plt.ylabel('mpg_avg')
plt.show()
print('There is an increasing trend in miles per gallon on an average from the year 1970 to 1982')
```

```python
plt.scatter(x=obj['mpg'],y=obj['horsepower'])
plt.xlabel('mpg')
plt.ylabel('horsepower')
plt.show()
print(' There is a decreasing trend in horsepower with increase in miles per gallon')
```

```python
plt.bar(obj['origin'],obj['mpg'],width=0.5)
plt.xlabel('origin')
plt.ylabel('mpg')
plt.show()
print('With increase in the value of origin average mpg is increasing. i.e average mpg of origin3>avergae mpg of origin2>\naverage mpg of origin 1')
```

```python
cylinder_group=obj.groupby(['cylinders'])
for name,group in cylinder_group:
    plt.boxplot(cylinder_group['mpg'].get_group(name).values)
    plt.xlabel('cylinder'+str(name))
    plt.ylabel('mpg')
    plt.show()
print(' As the number of cylinders increase miles per gallon decrease.\n Miles per gallon is highest for cylinder 4')
```
